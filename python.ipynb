```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import time
import collections
%matplotlib inline
```

## Functions

Phase $\Phi$:
$$
\Phi=\frac{t-t_0}{T}-\left \lfloor \frac{t-t_0}{T} \right\rfloor
$$

About `dual_sort`:

1. Create a dictionary, $d$, that functionally relates $x_i\mapsto y_i$
2. Define $xs=sorted(x)$.
3. Define $ys_i=d(xs_i)$
4. Returns $xs$, $ys$

`String_sum` calculates the *string-length* of $x$,$y$. I mean:
$$
\sum_0^{n-1} \sqrt{(x_{i+1}-x_i)^2+(y_{i+1}-y_i)^2}
$$

```python
Vfloat=np.vectorize(float)
Vround=np.vectorize(round)

def phase(t,t_0,T):
    return ((t-t_0)/T)-(Vround((t-t_0)/T))

def dual_sort(x,y):
    d=dict(zip(x,y))
    sx=sorted(x)
    sy=np.zeros(len(sx))
    for i in range(len(sx)):
        sy[i]=(d[sx[i]])
    return sx,sy

def string_sum(x,y):
    a=0
    for i in range(len(x)-1):
        a+=np.sqrt((x[i+1]-x[i])**2+(y[i+1]-y[i])**2)
    return a
```

### Data

```python
start=time.time()

data = open( './data/OGLE-LMC-CEP-0001.dat', 'r' )
lines = np.array(data.readlines( ))
splitted=np.core.defchararray.rsplit(lines)
HJD_string = np.zeros(len(splitted))
Magnitude_string = np.zeros(len(splitted))
error_string = np.zeros(len(splitted))

for i in range(int(len(splitted))):
    HJD_string[i]=(splitted[i][0])
    Magnitude_string[i]=(splitted[i][1])
    error_string[i]=(splitted[i][2])

HJD=Vfloat(HJD_string)
Mag=Vfloat(Magnitude_string)
err=Vfloat(error_string)

end=time.time()
print(end-start)
print(len(HJD))
```

```python
start=time.time()

p=np.arange(0.1,1,0.00002)
print(len(p)) 
sums=np.zeros(len(p))
q,w=0,len(HJD)
count=0
for i in range(len(p)):
    ph=phase(HJD,max(HJD[q:w]),p[i])
    sph,sm=dual_sort(ph,Mag[q:w])
    sums[i]=string_sum(sph,sm)
    count+=1
    if count%(int(len(p)/10.))==0:
        print(str(int(float(i)*100/len(p))+1)+"%")

end=time.time()
print(end-start)
```

```python
per=p[np.argwhere(sums==min(sums))[0][0]] 

fig1, ax = plt.subplots(figsize=(20, 8))
plt.scatter(p,sums,s=1,c='k') 
plt.plot([min(p),max(p)],[min(sums),min(sums)]) 
plt.xlabel("Period",fontsize=20)
plt.ylabel("String-Length-Sum",fontsize=20)
plt.axis([min(p),max(p),min(sums)*0.9,max(sums)])
plt.scatter([per],[min(sums)]) 

print(min(sums))
print(per)
```

```python
ph=phase(HJD,HJD[np.argwhere(Mag==min(Mag))[0][0]],per)
plt.scatter(ph,Mag,s=1)
```

3 loops $\Rightarrow$ divide by 3

```python
ph=phase(HJD,HJD[np.argwhere(Mag==min(Mag))[0][0]],per/3)
plt.scatter(ph,Mag,s=1)
period=per/3 #definitivo
```

Then, concatenate.

```python
final_phase=np.concatenate((ph,ph+1,ph+2))
final_magnitude=np.concatenate((Mag,Mag,Mag))
```
